```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir("Data/Dinesh_Tweets/") if isfile(join("Data/Dinesh_Tweets/", f))]
onlyfiles
```

```python
import pandas as pd
frames=[]
for i in onlyfiles:
    df=pd.read_csv(i)
    frames.append(df)
resulting_data=pd.concat(frames)
```
